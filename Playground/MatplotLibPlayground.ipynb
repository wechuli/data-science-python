```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
#
x = np.arange(11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Basic line plot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Figure x and x^2')
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r--') #dotted line
plt.subplot(1,3,2)
plt.plot(y,x,'g')
plt.subplot(1,3,3)
plt.plot(x,y,'g*') #star line
plt.plot(y,x,'r*-')#star and dotted line
plt.grid()

plt.tight_layout()
```

<h2>Object Oriented Plotting</h2>

```python
#createe a firgure plot and call 

fig1 = plt.figure() #empy canvas
axes = fig1.add_axes([0.1,0.1,0.9,0.9]) #add an axis left,bottom,width,height
axes.plot(x,y) #plot the data
axes.set_xlabel('X') #set x label
axes.set_ylabel('Y') #set y label
axes.set_title('title') #set title
```

```python
#Inset plot
fig2 = plt.figure()
axes_main = fig2.add_axes([0.1,0.1,0.9,0.9])
axes_inset = fig2.add_axes([0.3,0.5,0.3,0.4])
axes_main.plot(x,y,'b-')
axes_inset.plot(y,x,'r*-')
axes_main.set_xlabel('Main X')
axes_main.set_ylabel('Main Y')
axes_main.set_title('Main title')

axes_inset.set_xlabel('Inset X')
axes_inset.set_ylabel('Inset Y')
axes_inset.set_title('Inset Title')
```

Figure and a Set of Subplots

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,sharex=True,sharey=True)
plt.tight_layout()
```

```python
fig,axes = plt.subplots()
axes.plot(x,y,'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig1,axes1 = plt.subplots(nrows=2,ncols=1)
fig2,axes2 = plt.subplots(nrows=1,ncols=2)
```

```python
type(axes1),type(axes2)
```

```python
for ax in axes1:
    ax.plot(x,y,'b*-')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Title')
fig1.tight_layout()    
fig1
```

```python
axes2[0].plot(x,y,'y-*')
axes2[1].plot(x,y,'b-*')

axes2[0].set_xlabel('X')
axes2[0].set_ylabel('Y')
axes2[0].set_title('Title')

axes2[1].set_xlabel('X')
axes2[1].set_ylabel('Y')
axes2[1].set_title('Title')

fig2.tight_layout()
fig2
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='x**2')
axes.plot(x,x**3,label='x**3')
axes.legend(loc=0)
```

```python
#control how the line will look
fig,axes = plt.subplots()
axes.plot(x,x+1,'r.-')
axes.plot(x,x+3,'g--')
```

```python
#control how the line will look
fig,axes = plt.subplots()
axes.plot(x,x+1,color='blue',alpha=0.4)
axes.plot(x,x+2,color='#8B008B')
axes.plot(x,x+3,color='#FF8C00')
```

```python
# We can play around with the appearance of the plot
fig,axes = plt.subplots(figsize=(10,8))
axes.plot(x,y,color='violet',lw=4,ls='-',marker='s',markersize=9,markerfacecolor='blue',markeredgewidth=4,markeredgecolor='red')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(10,3))
axes[0].plot(x,y)
axes[0].set_title('default')

axes[1].plot(x,y)
axes[1].axis('tight')
axes[1].set_title('tight')

axes[2].plot(x,y)
axes[2].set_title('custom')
axes[2].set_xlim([1,4])
axes[2].set_ylim([0,50])
```
