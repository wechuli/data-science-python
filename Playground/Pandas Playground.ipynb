```python
import pandas as pd
import numpy as np
```

```python
my_label = ['x','y','z']
my_data = [100,200,300]
```

```python
pd.Series(data=my_data)
```

```python
pd.Series(data=my_data,index=my_label)
```

```python
#we can skip the labels 
pd.Series(my_data,my_label)
```

```python
#lets make a series from a numpy array
my_array = np.array(my_data)
```

```python
pd.Series(my_array,index=my_label)
```

<h2>Using Dictionaries<h2>

```python
my_dic = {'Paul':25,'Brian':27,'June':22,27:19}
```

```python
pd.Series(my_dic)
```

```python
for keys in my_dic.keys():
    #print(my_dic[key])
    print(keys)
```

```python
#Series can hold strings as well
pd.Series(my_dic.keys())
```
