<h2>Lambda Expressions<h2>

```python
#normal
def square(num):
    return num**2
```

```python
square(3)
```

```python
#Lambda version
lambda num:num**2
```

Map

```python
x = [1,2,3,4,5,6]
y=[]
for num in x:
    y.append(num**2)
print(y)
```

```python
result = map(square,x)
```

```python
list(result)
```

Using Lambda and Map

```python
mapresult = map(lambda num:num**3,x)
```

```python
list(mapresult)
```

```python
a=[1,2,3,4,5]
b=[3,4,35,5,3,2]
c=[-3,-3,56,43]
list(map(lambda x,y,z:x+y+z,a,b,c))
```

Filter Method

```python
list(filter(lambda num:num%2 == 0,range(1,10)))
```
