```python
import numpy as np
import pandas as pd
```

```python
data_dic = {'A':[1,2,np.nan,4,np.nan],
            'B':[np.nan,np.nan,np.nan,np.nan,np.nan],
            'C':[11,12,13,14,15],
            'D':[16,np.nan,18,19,20]}
df = pd.DataFrame(data_dic) # dataframe from a dic.
```

```python
df
```

```python
#we can ask if the data has null
df.isnull()
```

```python
df.notnull() #return true where not null
```

```python
df['A'].sum() #when taking the sum, nan is considered zero
```

```python
df['A'].mean() #when taking the mean, nan is ignored
```

```python
df.dropna(thresh=3)
```

```python
#Dropna
df.dropna(axis=1)
```

```python
#we can fill na
df.fillna(value='filled')
```

```python
df  #without the inplace option set to true, the dataframe does not change
```

```python
#fill with mean
df.fillna(value=df['A'].mean())
```

```python
#forward fill
df.fillna(method="ffill")
```

```python
#back fill
df.fillna(method='bfill')
```
