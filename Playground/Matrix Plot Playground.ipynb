```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

plt.hist(x='passengers',data=flights)

<h2>Heat Map</h2>

```python
#For heat map to work we need to make our data into a matrix format
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(),cmap='coolwarm',annot=True)
```

```python
ptf = flights.pivot_table(values='passengers', index='month', columns='year')
```

```python
ptf
```

```python
sns.heatmap(ptf,cmap='coolwarm',linecolor='white',linewidths=1)
```

<h2>Cluster Map</h2>

```python
sns.clustermap(ptf,cmap='coolwarm',standard_scale=1)
```
