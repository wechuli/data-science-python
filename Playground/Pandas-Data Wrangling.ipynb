<h2>Combining and Merging</h2>

```python
d1 = {'key':['a','b','c','d','e'],'A1':range(5),'B1':range(5,10)}
d2 = {'key':['a','b','c'],'A2':range(3),'B2':range(3,6)}
```

```python
d1
```

```python
d2
```

```python
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how='inner',on='key')
```
