```python
def compare():
    a = input('Insert number with variable name a: ')
    b = input('Insert number with variable name b: ')
    if a > b:
        print('{a} is greater than {b}'.format(a=a,b=b))
    elif a == b:
        print('{a} is equal to {b}'.format(a=a,b=b))
    else:
        print('{a} is smaller than {b}'.format(a=a,b=b))
```

```python
compare()
```
