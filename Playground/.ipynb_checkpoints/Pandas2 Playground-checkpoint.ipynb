```python
import pandas as pd
import numpy as np
```

```python
index = ['r1','r2','r3','r4','r5','r6','r7','r8','r9','r10']
cols = ['c1','c2','c3','c4','c5','c6','c7','c8','c9','c10']
array_2d = np.arange(0,100).reshape(10,10)
```
