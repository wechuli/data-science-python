<h2>Facet Grid </div>

```python
tips.head()
```

```python
tips['day'].nunique(),tips['time'].nunique(),tips['smoker'].nunique()
```

```python
fgrid = sns.FacetGrid(data=tips,col='time',row='smoker')
```

```python
fgrid = sns.FacetGrid(data=tips,col='time',row='smoker')
fgrid.map(plt.hist,'total_bill')
```

```python
fgrid = sns.FacetGrid(data=tips,col='time',row='smoker',hue='sex')
fgrid.map(plt.scatter,'total_bill','tip').add_legend()
```

<h2>Pair Grid</h2>

```python
iris.head()
```

```python
sns.PairGrid(data=iris)
```

```python
pairG = sns.PairGrid(data=iris,hue='species')
pairG.map(plt.scatter).add_legend()
```

```python
pairG = sns.PairGrid(data=iris,hue='species')
pairG.map_diag(plt.hist).add_legend()
pairG.map_offdiag(plt.scatter).add_legend()
```

```python
pair_g = sns.PairGrid(iris)
pair_g.map_diag(sns.distplot)
pair_g.map_upper(plt.scatter)
pair_g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(data=iris,hue='species',palette='rainbow')
```

<h2>Joint Grid</h2>

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
g = sns.JointGrid(x='total_bill',y='tip',data=tips)
g.plot(sns.regplot,sns.distplot)
```
