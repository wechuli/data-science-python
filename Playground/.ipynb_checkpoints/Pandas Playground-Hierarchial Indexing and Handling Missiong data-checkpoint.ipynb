<h2>Hierarchial Indexing</h2>

```python
import numpy as np
import pandas as pd
```

```python
index = [['a','a','a','b','b','b','c','c','d','d'],[1,2,3,1,2,3,1,2,1,2]]
```

```python
index
```

```python
ser = pd.Series(np.random.randn(10),index=index)
```

```python
ser
```

```python
ser['a']
```

```python
 #we can move to a more specific value
ser['a'][2]
```

```python
df = pd.DataFrame(np.arange(12).reshape((4,3)),index=[['a','a','b','b'],[1,2,1,2]],columns=['AB','ON','BC'])
```

```python
df
```

```python
df['AB']
```

```python
df['BC']
```

```python
df.loc['a']
```

```python
df.loc['b'].loc[2]
```

```python
df
```

```python
#to grab a specific value
df.loc['a'].loc[1]['BC']
```

```python
df.index.names
```

```python
df.index.names = ['L1','L2']  #we can name the multilevel index
```

```python
df.index.names
```

```python
df
```

```python
df.xs('a')
```

```python
df.xs(key=1,level='L2')
```
