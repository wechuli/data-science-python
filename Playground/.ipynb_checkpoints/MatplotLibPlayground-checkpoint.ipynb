```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
#
x = np.arange(11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Basic line plot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Figure x and x^2')
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r--') #dotted line
plt.subplot(1,3,2)
plt.plot(y,x,'g')
plt.subplot(1,3,3)
plt.plot(x,y,'g*') #star line
plt.plot(y,x,'r*-')#star and dotted line

plt.tight_layout()
```
