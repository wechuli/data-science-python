```python
import matplotlib as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips.describe()
```

```python
tips.head()
```

<h2>Strip Plot</h2>

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex')
```

<h2>Swarm Plot</h2>

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.swarmplot(x='total_bill',y='day',data=tips,hue='time')
```

<h2>Box Plot </h2>

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

<h2>Violin Plot</h2>

```python
sns.violinplot(x='day',y='total_bill',data=tips) #allows us to plot all the components that represent the actual data set
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

<h2>Bar Plot</h2>

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='sex') #mean will be used as the default aggregator
```

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='sex',estimator=np.std) #using the standard deviation as the estimator
```

```python
#count plot plots the number of occurence of a categorical variable
sns.countplot(x='day',data=tips,hue='sex')
```

```python
sns.countplot(x='size',data=tips)
```

<h2>Point Plot</h2>

```python
sns.pointplot(x='day',y='total_bill',data=tips,hue='sex') 
```

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='sex',alpha=0.5) 
sns.pointplot(x='day',y='total_bill',data=tips,hue='sex') 
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black',size=3)
```

<h2>FactorPlot</h2>

```python
sns.factorplot(x='sex',y='total_bill',data=tips,hue='smoker',kind='box') # plot any plot by adding kind
```

```python
sns.factorplot(x='day',y='total_bill',hue='smoker',data=tips,kind='swarm',col='time') #we can plot the various columns each on its plot
```

```python
sns.factorplot(x='day',y='total_bill',hue='smoker',data=tips,kind='box',col='day',aspect=0.5,size=4)
```
