```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

**Do the following tasks:**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.set_context('notebook')
sns.set_style('white')

titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
tips = sns.load_dataset('tips')
tips.head()
```

**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.boxplot(data=tips,x='day',y='tip',hue='sex')
```

**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hex')
```

**Recreate the distplot given below. Use titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
titanic.head()
```

```python
sns.distplot(a=titanic['fare'],kde=False,bins=20,color='green')
```

**Recreate the boxplot below using titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.boxplot(data=titanic,x='class',y='age',hue='alive',palette='rainbow')
```

**Recreate the plot below using titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.jointplot(data=titanic,x='fare',y='age')
```

**Recreate the plot below. Use titanic dataset. ** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.boxplot(data=titanic,x='class',y='age')
```

**Recreate the swarm plot using titanic dataset, use factorplot** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.swarmplot(data=titanic,x='class',y='age')
```

**Can you plot how many male and female were on titanic? Separate them based on their class.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.countplot(x='sex',data=titanic,hue='class')
```

**Recreate the plot below. Use iris dataset. Hint: PairGrid.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend()
```

**Recreate the plot below. Use titanic dataset. Hint: FacetGrid.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
g = sns.FacetGrid(data=titanic,col='sex', hue='alive')
g.map(plt.hist,'age',alpha=0.5)
g.add_legend()
```
