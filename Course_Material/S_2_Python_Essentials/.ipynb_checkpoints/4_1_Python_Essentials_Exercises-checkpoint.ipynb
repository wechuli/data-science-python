```python
# To avoid overwriting the output, please code here 
def findCal(sentence):
    splitS=sentence.lower().split()
    return 'calgary' in splitS

findCal('Calgary is a city in Alberta')
```

**13. Write a function to return square of numbers from 1 to 5, use range() method **

```python
# To avoid overwriting the output, please code here 
def square(numbers):
    result = []
    for number in numbers:
        result.append(number**2)
    return result

square(range(1,9))
```

```python
square(range(1,6))
```

**14. Write lambda expression for the function above, use map() and range()**

```python
# To avoid overwriting the output, please code here 
list(map(lambda x:x**2,range(1,9)))
```
